```python
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq
import numpy as np
```

```python
url = rq.get("http://www.statiz.co.kr/stat.php?opt=0&sopt=0&re=0&ys=2022&ye=2022&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR_ALL_ADJ&o2=TPA&de=1&lr=0&tr=&cv=&ml=1&sn=360&si=&cn=")
soup = BeautifulSoup(url.content, 'html.parser')
```

```python
temp = soup.find('table',id='mytable')
temp = temp.find_all('tr')
temp = [re.text.strip() for re in temp]
```

```python
temp = pd.Series(temp)
```

```python
temp
```

```python
temp = temp[~temp.str.match("[순W]")].reset_index(drop=True)
```

```python
temp
```

```python
temp = temp.apply(lambda x: pd.Series(x.split(" ")))
```

```python
temp
```

```python
temp = temp.replace('',np.nan).dropna(axis=1)
```
